```python
query_job_info=''' select distinct'''+year+'as App_year,'+''' job_title,job_education,work_postal_code,
sum(experience_num_months) as experience_num_months,sum(training_num_months) as training_num_months
from job_info group by job_title,job_education,work_postal_code '''
jobinfo = pd.read_sql(query_job_info,connection)
jobinfo.columns = jobinfo.columns.str.upper()
```

```python
query_application='''select'''+year+'as App_year,'+''' us_economic_sector
,case_status
,class_of_admission
,country_of_citizenship
,date_add(date_add(LAST_DAY(case_received_date), interval 1 DAY), interval - 1 MONTH) AS case_received_month
,date_add(date_add(LAST_DAY(decision_date), interval 1 DAY), interval - 1 MONTH) decision_month
,count(application_case_number) AS application_case_numbers
,sum(wage_offer_from) AS wage_offer_from
,sum(wage_offer_to) AS wage_offer_to
,sum(wage_offer_unit_of_pay_9089) AS wage_offer_unit_of_pay_9089
FROM application
GROUP BY us_economic_sector
,case_status
,class_of_admission
,country_of_citizenship
,date_add(date_add(LAST_DAY(case_received_date), interval 1 DAY), interval - 1 MONTH)
,date_add(date_add(LAST_DAY(decision_date), interval 1 DAY), interval - 1 MONTH)  '''
application = pd.read_sql(query_application,connection)
application.columns = application.columns.str.upper()
```

```python
application.replace(r'', np.NaN)
application[[         
'US_ECONOMIC_SECTOR','CASE_STATUS','CLASS_OF_ADMISSION','COUNTRY_OF_CITIZENSHIP'
]]=application[[
      'US_ECONOMIC_SECTOR','CASE_STATUS','CLASS_OF_ADMISSION','COUNTRY_OF_CITIZENSHIP' ]].fillna('Unknown')
```

```python
application[[         
    'CASE_RECEIVED_MONTH','DECISION_MONTH'
]]=application[[
        'CASE_RECEIVED_MONTH','DECISION_MONTH' ]].fillna('1900-01-01')
```

```python
application[[         
    'APPLICATION_CASE_NUMBERS','WAGE_OFFER_FROM','WAGE_OFFER_TO','WAGE_OFFER_UNIT_OF_PAY_9089'
]]=application[[
         'APPLICATION_CASE_NUMBERS','WAGE_OFFER_FROM','WAGE_OFFER_TO','WAGE_OFFER_UNIT_OF_PAY_9089'
]].fillna(0.00)
```

```python
application
```

```python
table_name = 'APPLICATION'
write_pandas(
            conn=conn,
            df=application,
            table_name=table_name,
            database=database1,
            schema=schema1
        )
```
